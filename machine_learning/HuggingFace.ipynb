```python
from transformers import pipeline
```

```python
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
texto = 'Transformers are a powerful deep learning architecture designed for processing sequential data, such as natural language, by utilizing a "self-attention" mechanism to evaluate the relevance of all other tokens in a sequence simultaneously'
```

```python
resumo = summarizer(texto, max_length=30,
                    min_length=5,
                    do_sample=False)
```

```python
print(resumo)
```

```python
classifier = pipeline("sentiment-analysis", top_k=None)  # ou "text-classification"
```

```python
texto2 = [
    "I loved this product, it works great!",
    "The service was terrible and very slow."
]
```

```python
resultados = classifier(texto2)
```

```python
print(resultados)
```

```python
translator = pipeline(
    "translation",
    model="facebook/nllb-200-distilled-600M",
    src_lang="eng_Latn",
    tgt_lang="por_Latn",
)
```

```python
out = translator("Transformers are changing natural language processing.", max_length=80)
print(out[0]["translation_text"])
```

```python
texto3 = "Transformers are changing natural language processing."
print(translator(texto3)[0]["translation_text"])
```
